```python
# Gradient Descent
import matplotlib.pyplot as plt
import numpy as np

X = np.array([[1, 1], [1, 2], [1, 3]])
x_vals = X[:, 1] # actual values, removes the bias
# print(x_vals)
y = np.array([1, 2, 3])
plt.plot(x_vals, y, color="blue", label="actual")
alpha, iterations = 0.01, 1000
# Rows represent observations
# Columns represent features
# scalar regression slope = y = mx+c but for vector = y = X @ theta (theta consists of both m and c) theta[0] = c, theta[1] = m
m, n = X.shape

theta = np.zeros(n)  # 2
for _ in range(iterations):
    y_pred = X @ theta  # y = mx + c and theta starts with 0 vals
    error = y_pred - y
    slope = 1 / m * (X.T @ error)  # gives partial derivatives of theta.
    theta = theta - slope * alpha

    # print(np.round(theta, 4).tolist())

print("Final theta:", np.round(theta, 4))
plt.plot(x_vals, y_pred, color="red", label="predicted")
plt.legend()
plt.title("Gradient Desc")
plt.grid(True)
plt.show()

# theta is a 2-element vector: [intercept, slope]
# x_new must also be 2 elements: [1, x_value]
# So x @ theta performs: slope calculation
def predict(x, theta):
    return x @ theta

x_new = np.array([1, 4]) # # bias + new input value
y_new = predict(x_new, theta)
print(y_new) # should print close to 4
```
